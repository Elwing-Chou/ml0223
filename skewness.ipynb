```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew

p = np.random.normal(10, 3, 10000)
p = 1 + p[p >= 0]
print(skew(p))
sns.distplot(p)
```

```python
p2 = p ** 4
print(skew(p2))
sns.distplot(p2)
```

```python
p2 = np.power(1.2, p)
print(skew(p2))
sns.distplot(p2)
```

```python
p = np.random.normal(10000, 5000, 10000)
p = 1 + p[p >= 0]

p2 = p ** (1 / 5)
print(skew(p2))
sns.distplot(p2)
```
